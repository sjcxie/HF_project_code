# Add "total_adms" attribute in baseline

We add the "total_adms" as the total number of admissions for that patient.

```python
for index, i in enumerate(admission.rID.unique()):
    numbers = admission[admission['rID']==i].shape[0]
    baseline = baseline.set_value(index,'total_adms',numbers)
```

# Save the updated two datasets again

```python
writer = pd.ExcelWriter('Advanced Therapy Potential Cohort.xlsx')
baseline.to_excel(writer, 'cohort baseline')
admission.to_excel(writer, 'cohort admission')
writer.save()
print('DataFrame is written successfully to Excel File.')
```
