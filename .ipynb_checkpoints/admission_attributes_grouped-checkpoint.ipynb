```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

pd.options.display.max_rows = 400
pd.options.display.max_columns = 50
```

```python
%cd ~/desktop/HF_Research/datasets/processed

admission = pd.read_excel('Filtered Attributes Admission.xlsx', 
                          'NA20_filtered_admission')
```

```python
baseline = pd.read_excel('Filtered Attributes Baseline.xlsx', 
                          'NA30_filtered_baseline')
```

# Update admission dataset
remove some unnecessary attributes

```python
diagnosis_codes = ['PRIMARY_DIAGNOSIS_CODE', 'PRIMARY_DIAGNOSIS_CODE_TYPE',
       'PRIMARY_DIAGNOSIS_DESCRIPTION', 'SECONDARY_DIAGNOSIS_CODE',
       'SECONDARY_DIAGNOSIS_CODE_TYPE', 'SECONDARY_DIAGNOSIS_DESCRIPTION',
       'TERTIARY_DIAGNOSIS_CODE', 'TERTIARY_DIAGNOSIS_CODE_TYPE',
       'TERTIARY_DIAGNOSIS_DESCRIPTION', 'QUATERNARY_DIAGNOSIS_CODE',
       'QUATERNARY_DIAGNOSIS_CODE_TYPE', 'QUATERNARY_DIAGNOSIS_DESCRIPTIO',
       'QUINARY_DIAGNOSIS_CODE', 'QUINARY_DIAGNOSIS_CODE_TYPE',
       'QUINARY_DIAGNOSIS_DESCRIPTION']
```

```python
self_created = ['admission_times', 'last_adm',
       'readmission_days_btw', 'readmit_30_label', 'inhospital_ctb_label',
       'ctb_label']
```

```python
admission.drop(columns=diagnosis_codes, inplace=True)
admission.drop(columns=self_created, inplace=True)
```

```python
writer = pd.ExcelWriter('Filtered Attributes Admission.xlsx')
admission.to_excel(writer, 'NA2030_filtered_admission', index=False)
writer.save()
print('DataFrame is written successfully to Excel File.')
```

# Combine baseline and admission 

```python
baseline_df = pd.DataFrame(columns=baseline.columns)

baseline_df.drop(columns=['rID'],inplace=True)
```

```python
baseline_df
```

```python
for i in baseline.rID:
    num = admission.loc[admission['rID']==i].shape[0]
    index = baseline_df.shape[0]
    fill = baseline[baseline['rID']==i]
    fill.drop(columns=['rID'], inplace=True)
    baseline_df = baseline_df.append([fill]*num)
```

```python
baseline_df.reset_index(drop=True, inplace=True)
```

```python
combined = pd.concat([admission, baseline_df], axis=1)
```

```python
combined.shape
```

# Save the combined dataset

```python
%pwd
```

```python
writer = pd.ExcelWriter('Filtered Combined Dataset.xlsx')
combined.to_excel(writer, 'Filtered_combined', index=False)
writer.save()
print('DataFrame is written successfully to Excel File.')
```
